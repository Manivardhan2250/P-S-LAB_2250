```python
# prompt: Question 1: Simple Linear Regression
# Problem Statement: You are given data on individuals’ heights and their corresponding weights. Use Simple Linear
# Regression to predict a person’s weight based on height.
# Height (cm) Weight (kg)
# 150 50
# 155 52
# 160 56
# 165 58
# 170 62
# 175 66
# 180 70
# 185 74
# 190 78
# Tasks:
# • Fit a linear regression model.
# • Plot height vs weight along with the regression line.
# • Predict the weight of a person who is 172 cm tall.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given data
height = np.array([150, 155, 160, 165, 170, 175, 180, 185, 190]).reshape((-1, 1))
weight = np.array([50, 52, 56, 58, 62, 66, 70, 74, 78])

# Fit the linear regression model
model = LinearRegression()
model.fit(height, weight)

# Predict weight for a person with height 172 cm
height_to_predict = np.array([[172]])
predicted_weight = model.predict(height_to_predict)

# Plot height vs weight and the regression line
plt.scatter(height, weight, color='blue', label='Data Points')
plt.plot(height, model.predict(height), color='red', label='Regression Line')
plt.scatter(height_to_predict, predicted_weight, color='green', label='Prediction')

plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()

print(f"Predicted weight for a person with height 172 cm: {predicted_weight[0]:.2f} kg")
```

```python
# prompt: Question 2: Multiple Linear Regression
# Problem Statement: You are given data about used cars including horsepower, mileage, and age. Use Multiple Linear
# Regression to predict the price of a car.
# Horsepower Mileage (mpg) Age (years) Price ($1000s)
# 120 35 5 20
# 150 30 4 25
# 130 33 3 27
# 160 28 2 30
# 110 38 6 18
# 140 32 4 24
# Tasks:
# • Train a multiple linear regression model.
# • Predict the price of a car with 135 horsepower, 31 mpg, and 3 years old.
# • Interpret the importance of each feature.

import numpy as np
from sklearn.linear_model import LinearRegression

# Given data
X = np.array([[120, 35, 5], [150, 30, 4], [130, 33, 3], [160, 28, 2],
              [110, 38, 6], [140, 32, 4]])
y = np.array([20, 25, 27, 30, 18, 24])

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Predict the price for a new car
new_car = np.array([[135, 31, 3]])
predicted_price = model.predict(new_car)

print(f"Predicted price for the new car: ${predicted_price[0]:.2f} thousands")

# Interpret feature importance (coefficients)
print("Feature Importance (Coefficients):")
print(f"Horsepower: {model.coef_[0]:.2f}")
print(f"Mileage: {model.coef_[1]:.2f}")
print(f"Age: {model.coef_[2]:.2f}")

# Intercept
print(f"Intercept: {model.intercept_:.2f}")
```

```python
# prompt: Question 3: Polynomial Regression
# Problem Statement: You are given data about how much sunlight a plant receives daily and its height. The relationship
# is non-linear. Use Polynomial Regression (degree 2) to model the data and predict plant growth.
# 1
# Sunlight (hours/day) Plant Height (cm)
# 1 5
# 2 9
# 3 15
# 4 23
# 5 30
# 6 32
# 7 31
# 8 29
# Tasks:
# • Transform the input features using a degree-2 polynomial.
# • Fit the polynomial regression model.
# • Plot the original data and fitted curve.
# • Predict plant height for 6.5 hours of sunlight.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Given data
sunlight = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape((-1, 1))
height = np.array([5, 9, 15, 23, 30, 32, 31, 29])

# Transform input features using degree-2 polynomial
poly = PolynomialFeatures(degree=2)
sunlight_poly = poly.fit_transform(sunlight)

# Fit the polynomial regression model
model = LinearRegression()
model.fit(sunlight_poly, height)

# Predict plant height for 6.5 hours of sunlight
sunlight_to_predict = np.array([[6.5]])
sunlight_to_predict_poly = poly.transform(sunlight_to_predict)
predicted_height = model.predict(sunlight_to_predict_poly)

# Generate points for plotting the fitted curve
sunlight_range = np.linspace(1, 8, 100).reshape(-1, 1)
height_pred = model.predict(poly.transform(sunlight_range))


# Plot the original data and fitted curve
plt.scatter(sunlight, height, color='blue', label='Data Points')
plt.plot(sunlight_range, height_pred, color='red', label='Polynomial Regression Curve')
plt.scatter(sunlight_to_predict, predicted_height, color='green', label='Prediction')

plt.xlabel('Sunlight (hours/day)')
plt.ylabel('Plant Height (cm)')
plt.title('Polynomial Regression')
plt.legend()
plt.grid(True)
plt.show()

print(f"Predicted plant height for 6.5 hours of sunlight: {predicted_height[0]:.2f} cm")
```
