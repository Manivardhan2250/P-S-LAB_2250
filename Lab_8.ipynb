```python
# prompt: You are given a dataset containing information about Advertising Budget, Sales Revenue, and Social Media Engagement.
# Advertising Budget	Sales Revenue	Social Media Engagement
# 1000	5000	200
# 2000	7000	250
# 3000	8000	300
# 4000	9500	400
# 1500	6000	220
# 2500	7500	280
# 3500	8700	350
# 1200	5500	210
# 4500	9900	450
# 5000	11000	500
# 1.	Create scatter plots for the following pairs of variables:
# o	Advertising Budget vs. Sales Revenue
# o	Social Media Engagement vs. Sales Revenue
# o	Advertising Budget vs. Social Media Engagement
# 2.	Calculate the Pearson correlation coefficient for each pair and interpret the results.
# 3.	Discuss whether increasing the advertising budget always leads to higher sales revenue. How does social media engagement impact sales?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Create a DataFrame from the provided data
data = {'Advertising Budget': [1000, 2000, 3000, 4000, 1500, 2500, 3500, 1200, 4500, 5000],
        'Sales Revenue': [5000, 7000, 8000, 9500, 6000, 7500, 8700, 5500, 9900, 11000],
        'Social Media Engagement': [200, 250, 300, 400, 220, 280, 350, 210, 450, 500]}
df = pd.DataFrame(data)

# 1. Create scatter plots
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.scatterplot(x='Advertising Budget', y='Sales Revenue', data=df)
plt.title('Advertising Budget vs. Sales Revenue')

plt.subplot(1, 3, 2)
sns.scatterplot(x='Social Media Engagement', y='Sales Revenue', data=df)
plt.title('Social Media Engagement vs. Sales Revenue')

plt.subplot(1, 3, 3)
sns.scatterplot(x='Advertising Budget', y='Social Media Engagement', data=df)
plt.title('Advertising Budget vs. Social Media Engagement')

plt.tight_layout()
plt.show()

# 2. Calculate Pearson correlation coefficients
pairs = [('Advertising Budget', 'Sales Revenue'),
         ('Social Media Engagement', 'Sales Revenue'),
         ('Advertising Budget', 'Social Media Engagement')]

for pair in pairs:
  correlation, p_value = pearsonr(df[pair[0]], df[pair[1]])
  print(f"Pearson correlation between {pair[0]} and {pair[1]}: {correlation:.2f}")
  print(f"P-value: {p_value:.3f}")
  if p_value < 0.05:
    print("The correlation is statistically significant.")
  else:
    print("The correlation is not statistically significant.")
  print("-" * 20)

# 3. Discussion
print("Discussion:")
print("Increasing the advertising budget generally leads to higher sales revenue, as indicated by the positive correlation.")
print("However, the scatter plots reveal some instances where sales do not increase proportionally with the budget, suggesting other factors at play.")
print("Social media engagement appears to have a strong positive correlation with sales revenue.  Higher engagement tends to correspond with higher sales.")
print("It is important to consider social media engagement as a potential factor for increased sales alongside advertising budget.")
```
