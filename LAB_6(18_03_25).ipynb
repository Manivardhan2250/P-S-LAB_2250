```python
# prompt: Generate a normal distribution of data points with a mean of 75 and a standard
# deviation of 10. Plot a histogram for the generated data. Analyze the symmetry and
# bell-shaped curve of a normal distribution and visualize its Probability Density
# Function (PDF).

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data from a normal distribution
mean = 75
std_dev = 10
data = np.random.normal(loc=mean, scale=std_dev, size=1000)

# Plot a histogram of the data
plt.hist(data, bins=30, density=True, alpha=0.6, color='b')

# Plot the PDF of the normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)

plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()

# Analyze the symmetry and bell-shaped curve
print("Mean:", np.mean(data))
print("Standard Deviation:", np.std(data))
print("The data should exhibit a symmetrical bell-shaped curve centered around the mean.")
```

```python
# prompt: The normalized distribution for a person’s body mass index (BMI) has a mean of 20
# and a standard deviation of 4. Calculates and visualizes the following probabilities for
# a randomly selected person having a BMI:
# I. greater than or equal to 23.
# II. less than or equal to 23.
# III. between 23 and 35.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


# Parameters of the normal distribution
mean = 20
std_dev = 4

# Calculate probabilities
prob_greater_equal_23 = 1 - norm.cdf(23, mean, std_dev)
prob_less_equal_23 = norm.cdf(23, mean, std_dev)
prob_between_23_35 = norm.cdf(35, mean, std_dev) - norm.cdf(23, mean, std_dev)


# Print the calculated probabilities
print("Probability of BMI greater than or equal to 23:", prob_greater_equal_23)
print("Probability of BMI less than or equal to 23:", prob_less_equal_23)
print("Probability of BMI between 23 and 35:", prob_between_23_35)


# Visualize the distribution and probabilities
x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 100)
y = norm.pdf(x, mean, std_dev)

plt.plot(x, y)
plt.fill_between(x, y, where=(x >= 23), color='red', alpha=0.5)
plt.fill_between(x, y, where=(x <= 23), color='blue', alpha=0.5)
plt.fill_between(x, y, where=(x >= 23) & (x <= 35), color='green', alpha=0.5)


plt.title('BMI Distribution with Probabilities')
plt.xlabel('BMI')
plt.ylabel('Probability Density')

plt.legend(['PDF', 'BMI >= 23', 'BMI <= 23', '23 <= BMI <= 35'])
plt.show()
```
