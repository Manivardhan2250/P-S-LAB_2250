```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Question 1: Analyzing Daily Customer Visits
np.random.seed(42)  # For reproducibility
N = 100  # Number of days
customers_per_day = np.random.randint(0, 11, N)  # Random values between 0 and 10

# Compute PMF
unique, counts = np.unique(customers_per_day, return_counts=True)
pmf = counts / N

# Compute CDF
cdf = np.cumsum(pmf)

# Plot PMF and CDF
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(unique, pmf, color='b', alpha=0.6, label='PMF')
ax2.plot(unique, cdf, color='r', marker='o', linestyle='-', label='CDF')
ax1.set_xlabel('Number of Customers')
ax1.set_ylabel('PMF', color='b')
ax2.set_ylabel('CDF', color='r')
ax1.set_title('PMF and CDF of Daily Customer Visits')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.show()
```

```python
# Question 2: Exam Pass Probability
N_students = 100
p_pass = 0.7

# (i) Probability that exactly 60 students pass
prob_60 = stats.binom.pmf(60, N_students, p_pass)

# (ii) Probability that at least 75 students pass
prob_at_least_75 = 1 - stats.binom.cdf(74, N_students, p_pass)

# (iii) Expected number of students passing
expected_pass = N_students * p_pass

# Print results
print(f"(i) Probability of exactly 60 students passing: {prob_60:.5f}")
print(f"(ii) Probability of at least 75 students passing: {prob_at_least_75:.5f}")
print(f"(iii) Expected number of students passing: {expected_pass}")
```
